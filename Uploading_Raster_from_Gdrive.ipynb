```
import pandas as pd
import numpy as np

df = pd.read_csv('TRIAL_RASTER.csv')
raster = df.to_numpy()

# Print the NumPy array
print(raster)
```

```
df.head()
```

```
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(16,20))

rastplot = plt.imshow(raster, cmap='viridis')

# plt.getp(rastplot)

# Create a plot
rastplot = plt.imshow(raster)

# Convert the plot to a NumPy array
rastplot_array = np.array(rastplot)

# Save the plot
plt.imsave('Raster.jpg', raster)

plt.show()
```

```
from sklearn.decomposition import PCA

# Create a PCA object
pca = PCA()

# Fit the PCA model to the data
pca.fit(raster)

# Get the principal components
PCs = pca.components_

plt.plot(PCs[0,:])
```

```
PCs.shape

PCs_Norm = np.zeros(PCs.shape)

# normalize each PC individually for visualization
for k in range(PCs.shape[0]):
  PCs_Norm[k,:] = (PCs[k,:]-np.min(PCs[k,:]))/(np.max(PCs[k,:])-np.min(PCs[k,:]))

plt.imshow(PCs_Norm, vmin=0, vmax=1)
plt.show()
```
